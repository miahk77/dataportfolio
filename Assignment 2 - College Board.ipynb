```python
#convert all participation rates for ACT and SAT to floats using a function that removes the % symbol. 

def strippercentage(str):
    return str.replace("%","")
```

```python
#applying function to ACT participation rates

combinedactsat["Act_Participation_2017"]= combinedactsat["Act_Participation_2017"].apply(strippercentage)
combinedactsat["Act_Participation_2018"]= combinedactsat["Act_Participation_2018"].apply(strippercentage)
combinedactsat["Act_Participation_2019"]= combinedactsat["Act_Participation_2019"].apply(strippercentage)
```

```python
#in order to do the same for the SAT participation rates, need to fillna in overall dataset with 0

combinedactsat.fillna("0", inplace=True)
```

```python
combinedactsat["Sat_Participation_2017"]= combinedactsat["Sat_Participation_2017"].apply(strippercentage)
combinedactsat["Sat_Participation_2018"]= combinedactsat["Sat_Participation_2018"].apply(strippercentage)
combinedactsat["Sat_Participation_2019"]= combinedactsat["Sat_Participation_2019"].apply(strippercentage)
```

```python
#convert all total composite and total rates for ACT and SAT to floats. 

combinedactsat["Act_Participation_2019"] = pd.to_numeric(combinedactsat["Act_Participation_2019"], downcast="float")
combinedactsat["Act_Participation_2017"] = pd.to_numeric(combinedactsat["Act_Participation_2017"], downcast="float")
combinedactsat["Act_Participation_2018"] = pd.to_numeric(combinedactsat["Act_Participation_2018"], downcast="float")
combinedactsat["Sat_Participation_2017"] = pd.to_numeric(combinedactsat["Sat_Participation_2017"], downcast="float")
combinedactsat["Sat_Participation_2018"] = pd.to_numeric(combinedactsat["Sat_Participation_2018"], downcast="float")
combinedactsat["Sat_Participation_2019"] = pd.to_numeric(combinedactsat["Sat_Participation_2019"], downcast="float")
combinedactsat["PctGE21"] = pd.to_numeric(combinedactsat["PctGE21"], downcast="float")
```

#### CHART W/TWO Y AXIS

```python
fig, axes = plt.subplots(1, 2, figsize=(25, 10))

State = combinedactsat["State"]

Act_Participation_2019 = combinedactsat["Act_Participation_2019"]

Sat_Participation_2019 = combinedactsat["Sat_Participation_2019"]

#plotting two charts
axes[0].plot(State,Act_Participation_2019)
axes[1].plot(State, Sat_Participation_2019);


# Instantiate a second axes that shares the same x-axis
fig, ax1 = plt.subplots(figsize=(25, 10))

ax2 = ax1.twinx()  
ax2.set_ylim(4, 1500);
```

### COMPARING SAT 2019 AND ACT 2019 PARTICIPATION RATES 

This chart provides some insight on how different states participation rates vary between SAT and ACT scores. Just by a first look, we can see that when ACT rates are higher in a given state, SAT scores are generally lower. 

```python
COLOR_ACT = "#69b3a2"
COLOR_SAT = "#3399e6"

#ax.twinx() to create twin axes 
fig, ax1 = plt.subplots(figsize=(30, 15))
ax2 = ax1.twinx()

#plotting the axes and changing line color 
ax1.plot(State, Act_Participation_2019, color = COLOR_ACT, lw=4,marker='o')
ax2.plot(State, Sat_Participation_2019, color = COLOR_SAT, lw=4, linestyle='--', marker='o');

#setting labels and formatting 
ax1.set_xlabel("STATE")
ax1.set_ylabel("ACT PARTICIPATION 2019", color =COLOR_ACT, fontsize=15)
ax2.set_ylabel("SAT PARTICIPATION 2019", color = COLOR_SAT,fontsize=15)
fig.suptitle("SAT AND ACT PARTICIPATION 2019", fontsize=20)

#changing the x axis to avoid overlapping - got this from https://pythonguides.com/matplotlib-x-axis-label/
plt.setp(ax1.get_xticklabels(), rotation=30, ha='right')

plt.show()
```
